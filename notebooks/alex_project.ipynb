```python
import pandas as pd
from pathlib import Path
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
acb_path = Path('../data_files/ACB.csv')
apha_path = Path('../data_files/APHA.TO.csv')
crbp_path = Path('../data_files/CRBP.csv')
cron_path = Path('../data_files/CRON.TO.csv')
hexo_path = Path('../data_files/HEXO.TO.csv')
tgod_path = Path('../data_files/TGOD.TO.csv')
tlry_path = Path('../data_files/TLRY.csv')
trst_path = Path('../data_files/TRST.TO.csv')
weed_path = Path('../data_files/WEED.TO.csv')
```

```python
acb_data = pd.read_csv(acb_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
apha_data = pd.read_csv(apha_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
crbp_data = pd.read_csv(crbp_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
cron_data = pd.read_csv(cron_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
hexo_data = pd.read_csv(hexo_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
tgod_data = pd.read_csv(tgod_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
tlry_data = pd.read_csv(tlry_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
trst_data = pd.read_csv(trst_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
weed_data = pd.read_csv(weed_path, parse_dates=True, index_col='Date', infer_datetime_format=True)
```

```python
##Concatenated data together with ticker names -Alex
top10_df = pd.concat([acb_data, apha_data, crbp_data, cron_data, hexo_data, tgod_data, tlry_data, trst_data, weed_data], axis="rows", join="inner")
```

```python
top10_df
```

```python
##Created a pivot table for closing data of individual stocks of the MJ index -Alex
top10_table = pd.pivot_table(top10_df, values=["Close"], index="Date", columns="Ticker")
top10_table
```

```python
top10_table_edit = top10_table.fillna(method = 'ffill')
```

```python
##Created a basic plot of closing data, will come back to it -Alex
top10_table_edit.plot(figsize=(20,10))
```

```python
daily_returns = top10_table.pct_change()
daily_returns.head()
```

```python
daily_std = daily_returns.std()
daily_std
```

```python
annual_std = daily_std * np.sqrt(252)
annual_std
```

```python
sharpe_ratios = (top10_table_edit.mean() * 252) / (top10_table_edit.std() * np.sqrt(252))
sharpe_ratios
```

```python
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios of Individual Stocks in MJ Index")
```

```python
corr_top10 = top10_table_edit.corr()
corr_top10
```
