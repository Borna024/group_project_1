```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import os
import seaborn as sns
%matplotlib inline
```

```python
#Creating dataframe of index based on alcohol stocks and tobacco stocks respectively
```

```python
# Import alcohol stock CSVs
abev_csv = Path("./data/ABEV.csv")
brew_csv = Path("./data/BREW.csv")
bud_csv = Path("./data/BUD.csv")
deo_csv = Path("./data/DEO.csv")
fmx_csv = Path("./data/FMX.csv")
sam_csv = Path("./data/SAM.csv")
stz_csv = Path("./data/STZ.csv")
tap_csv = Path("./data/TAP.csv")
wvvi_csv = Path("./data/WVVI.csv")


# Read CSVs for alcohol stocks
abev_df = pd.read_csv(abev_csv, index_col="date", infer_datetime_format=True, parse_dates=True)
brew_df = pd.read_csv(brew_csv, index_col="date", infer_datetime_format=True, parse_dates=True)
bud_df = pd.read_csv(bud_csv, index_col="date", infer_datetime_format=True, parse_dates=True)
deo_df = pd.read_csv(deo_csv, index_col="date", infer_datetime_format=True, parse_dates=True)
fmx_df = pd.read_csv(fmx_csv, index_col="date", infer_datetime_format=True, parse_dates=True)
sam_df = pd.read_csv(sam_csv, index_col="date", infer_datetime_format=True, parse_dates=True)
stz_df = pd.read_csv(stz_csv, index_col="date", infer_datetime_format=True, parse_dates=True)
tap_df = pd.read_csv(tap_csv, index_col="date", infer_datetime_format=True, parse_dates=True)
wvvi_df = pd.read_csv(wvvi_csv, index_col="date", infer_datetime_format=True, parse_dates=True)



# Import tobacco stock CSVs
mo_csv = Path("./data/MO.csv")
pm_csv = Path("./data/PM.csv")

# Read Tobacco CSVs
mo_df = pd.read_csv(mo_csv, index_col="date", infer_datetime_format=True, parse_dates=True)
pm_df = pd.read_csv(pm_csv, index_col="date", infer_datetime_format=True, parse_dates=True)


# import biotech csv
xbi_csv = Path("./data/XBI.csv")

# read biotech csv
xbi_df = pd.read_csv(xbi_csv, index_col="date", infer_datetime_format=True, parse_dates=True)


# import pharmaceutical csv
xph_csv = Path("./data/XPH.csv")

#read pharmaceutical csv
xph_df = pd.read_csv(xph_csv, index_col="date", infer_datetime_format=True, parse_dates=True)


#import MJ csv
mj_csv = Path("./data/MJETF.csv")

#read MJ csv

mj_df = pd.read_csv(mj_csv, index_col="date", infer_datetime_format=True, parse_dates=True)




```

```python
# combine CSVs into indexes
alcohol_index = pd.concat([abev_df, brew_df, bud_df, deo_df, fmx_df, sam_df, stz_df, tap_df, wvvi_df], axis=1 , join="inner")
alcohol_index.drop(columns = ["volume", "open", "high", "low"], inplace = True)
alcohol_index.columns = ["ABEV", "BREW", "BUD", "DEO", "FMX", "SAM", "STZ", "TAP", "WVVI"]
alcohol_index.head()
```

```python
# create tocacco index
tobacco_index = pd.concat([mo_df, pm_df], axis=1, join="inner")
tobacco_index.drop(columns = ["volume", "open", "high", "low"], inplace = True)
tobacco_index.columns = ["MO", "PM"]

tobacco_index.head()
```

```python
# biotech index
biotech_index = xbi_df
biotech_index.drop(columns = ["volume", "open", "high", "low"], inplace = True)
biotech_index.columns = ["XBI"]

biotech_index.head()
```

```python
#pharma index
pharma_index = xph_df
pharma_index.drop(columns = ["volume", "open", "high", "low"], inplace = True)
pharma_index.columns = ["XPH"]

pharma_index.head()
```

```python
# MJ index
mj_index = mj_df
mj_index.columns = ["MJ"]

mj_index.head()
```

```python
alcohol_weights = [1/9, 1/9, 1/9, 1/9, 1/9, 1/9, 1/9, 1/9, 1/9]
tobacco_weights = [1/2, 1/2]

weighted_alcohol_index = alcohol_index.dot(alcohol_weights)
weighted_tobacco_index = tobacco_index.dot(tobacco_weights)
```
